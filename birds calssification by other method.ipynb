```python
data = train_csv_file
def discrete(j, data):
    A = []
    c = 0
    B = {}
    D = []

    column_values = data[j]

    for i in column_values:
        if i not in A:
            A.append(i)
            c = c + 1
            B[i] = c
        D.append(B[i])

    return A

lable = discrete('labels', data)
print(lable.head())
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(train_csv_file,
                                      test_size = 0.3,
                                      random_state = 42)
```

```python
train_df.head()
```

```python
# train

X_train_path = train_df['filepaths'].tolist()
y_train = train_df['class id'].astype(int).tolist()
y_train = np.array(y_train)
X_train = []
for image_path in X_train_path:
  full_path = os.path.join('/content/Upload/Upload/',image_path)
  image = cv2.imread(full_path)[...,::-1]
  image = cv2.resize(image, (68, 68))
  #image = Image.open(full_path)
  #image = np.array(image)
  X_train.append(image)
```

```python
X_train = np.array(X_train)
print(X_train.shape)
X_train = X_train.reshape((len(X_train), 68, 68 ,3))
s = y_train.shape
print(s)
```

```python
X_train = np.array(X_train)
print("shape", X_train.shape)
```

```python
X_test_path = test_csv_file['filepaths'].tolist()
print(X_test_path[0])
X_test = []
for image_path in X_test_path:
  try:

    full_path = os.path.join('/content/Upload/Upload/',image_path)
    image = cv2.imread(full_path)[...,::-1]
    image = cv2.resize(image, (68, 68))
  #image = Image.open(full_path)
  #image = np.array(image)
  except TypeError as e:
    image = np.zeros((68,68,3), dtype=np.uint8)
  X_test.append(image)
```

```python
X_test = np.array(X_test)
print("shape", X_test.shape)
```

```python
# valid

X_val_path = test_df['filepaths'].tolist()
y_val = test_df['class id'].astype(int).tolist()
y_val = np.array(y_val)
X_val = []
for image_path in X_val_path:
  full_path = os.path.join('/content/Upload/Upload/',image_path)
  #image = Image.open(full_path)
  image = cv2.imread(full_path)[...,::-1]
  image = cv2.resize(image, (68, 68))
  X_val.append(image)
```

```python
X_val = np.array(X_val)
print("shape", X_val.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder

labels = np.array(y_train).reshape(-1, 1)
encoder = OneHotEncoder(sparse=False)
Y_train = encoder.fit_transform(labels)
print(Y_train[3000].shape)

labels_val = np.array(y_val).reshape(-1, 1)
encoder = OneHotEncoder(sparse=False)
Y_val = encoder.fit_transform(labels_val)
print(Y_val[1000].shape)
print(len(X_train))
print(Y_train.shape)
```

```python
model = Sequential()
model.add(Conv2D(32,3,padding="same", activation="relu", input_shape=(68,68,3)))
model.add(MaxPool2D())

model.add(Conv2D(32, 3, padding="same", activation="relu"))
model.add(MaxPool2D())

model.add(Conv2D(64, 3, padding="same", activation="relu"))
model.add(MaxPool2D())
model.add(Dropout(0.4))

model.add(Flatten())
model.add(Dense(128,activation="relu"))
model.add(Dense(524, activation="softmax"))

model.summary()
```

```python
opt = Adam(lr=0.000001)
#model.compile(optimizer = opt , loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False) , metrics = ['accuracy'])
#model.compile(optimizer = opt, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
history = model.fit(X_train,Y_train,epochs = 5 , validation_data = (X_val, Y_val))
```

```python
print(X_test.shape)
predicted_probs = model.predict(X_test)  # x_test is your input data
predicted_classes = np.argmax(predicted_probs, axis=1)
print(predicted_classes.shape)
```
