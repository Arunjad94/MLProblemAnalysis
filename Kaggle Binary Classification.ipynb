```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler as stdscl
import matplotlib.pyplot as plt
from google.colab import drive
import datetime
import seaborn as sns
from collections import Counter
import math
drive.mount('/content/drive')
path = '/content/drive/MyDrive/Assingment/Binary_Classification'
train = pd.read_csv(path + '/train.csv')
test = pd.read_csv(path + '/test.csv')
```

```python
print(train.columns)
corr = train.corr()['label']
print(corr.sort_values(ascending=False))
```

```python
#X_train = train.drop(columns=['label', 'Unnamed: 0', 'column_q', 'column_ab', 'column_t', 'column_z'])
X_train = train.drop(columns=['label','Unnamed: 0'])
y_train = train['label']
#X_test = test.drop(columns=['Unnamed: 0', 'column_q', 'column_ab', 'column_t', 'column_z'])
X_test = test.drop(columns=['Unnamed: 0'])
```

```python
def Binary_classificatio(X, y):
  model = LogisticRegression()
  model.fit(X, y)
  return model

def Performence(model, X, y):
  y_pred = model.predict(X)
  accuracy = metrics.accuracy_score(y, y_pred)
  f1 = f1_score(y, y_pred)
  return y_pred, accuracy, f1
```

```python
model = Binary_classificatio(X_train, y_train)
y_pred, accuracy, f1 = Performence(model, X_train, y_train)
print(y_pred, accuracy, f1)
```

```python
Prediction =  model.predict(X_test)
A, = Prediction.shape
Id = np.arange(A)
df = pd.DataFrame({'ID': Id, 'label': Prediction})
csv_file = '/content/drive/MyDrive/Assingment/Binary_Classification/Arun_Dilli_Submission.csv'
df.to_csv(csv_file, index=False)
```
